```python
# randomisation:
#   np.random.seed: enregistrer la valeur du random
#   .loc : utiliser pour mentrer le data entre crochet
#   np.random.permutation : créer un random de matrice aléatoire

np.random.seed(1)
paris_listings = paris_listings.loc[np.random.permutation(len(paris_listings))]
paris_listings
```

```python
# supprimer les virgules et les signes $ de price
paris_listings['price']=paris_listings['price'].str.replace('$','')
paris_listings['price']=paris_listings['price'].str.replace(',' , '')
paris_listings['price']=paris_listings['price'].astype('float')
paris_listings.head()
```

```python
def predict_price(new_listings):
    temp_df = paris_listings.copy()
    temp_df['distance'] = temp_df['accommodates'].apply(lambda x: np.abs(new_listings - x))
    temp_df= temp_df.sort_values('distance')
    nearest_neighbors = temp_df.iloc[0:5]['price']
    predicted_price = nearest_neighbors.mean()
    return(predicted_price)
predict_price(1)
```

### difision du tableau en train (6000 lignes) et test 

```python
train_df = paris_listings[:6000]
```

```python
test_df = paris_listings[6000:]
```

### nouvelle fonction de prédiction

```python
def predict_price(new_listings):
    df_temps = train_df.copy()
    df_temps['distance'] = df_temps['accommodates'].apply(lambda x: np.abs(x - new_listings))
    df_temps = df_temps.sort_values('distance')
    predict_price = df_temps.iloc[0:5]['price'].mean()
    return predict_price
```

###  application de la fonction dans le data test

```python
test_df["predicted_price"] = test_df["accommodates"].apply(lambda x:predict_price(x)) 
```

```python
test_df.head()
```

## calcul de la métrique MAE

```python
MAE=np.abs(test_df['price'] - test_df['predicted_price']).mean()
MAE
```

### calcul de MSE

```python
MSE=((test_df['price'] - test_df['predicted_price'])**2).mean()
MSE
```

### entrainement d'un autre model

```python
def predictionII(bedroom_number):
    df_temp = train_df.copy()
    df_temp['distance'] = df_temp['bedrooms'].apply(lambda x: np.abs(x - bedroom_number))
    df_temp = df_temp.sort_values('distance')
    predict_price = df_temp.iloc[0:5]['price'].mean()
    return predict_price
```

```python
test_df['predicted_price'] = test_df['bedrooms'].apply(lambda x: predictionII(x))
```

```python
MAE_II=np.abs(test_df['price'] - test_df['predicted_price']).mean()
MAE_II
```

```python
MSE_II=((test_df['price'] - test_df['predicted_price'])**2).mean()
MSE_II
```

## calcul de RMSE

```python
RMSE = np.sqrt(MSE_II)
RMSE
```
