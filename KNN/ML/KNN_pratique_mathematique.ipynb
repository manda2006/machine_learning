```python
import numpy as np
our_acc_value=3
first_giving_space_value = paris_listings.iloc[0]['accommodates']

first_distance = np.abs(our_acc_value-first_giving_space_value)
print(first_distance)
```

```python
## caclul de la distance euclidienne pour plusieurs donné
```

```python
new_listings=3
paris_listings['distance'] = paris_listings['accommodates'].apply(
lambda x: np.abs(new_listings-x))
print(paris_listings['distance'].value_counts())
```

```python
# randomisation:
#   np.random.seed: enregistrer la valeur du random
#   np.random.permutation : créer un random de matrice aléatoire

np.random.seed(1)
paris_listings = paris_listings.loc[np.random.permutation(len(paris_listings))]
paris_listings
```

```python
paris_listings = paris_listings.sort_values('distance')
```

```python
paris_listings['price'].head(10)
```

```python
# supprimer les virgules et les signes $ de price
paris_listings['price']=paris_listings['price'].str.replace('$','')
paris_listings['price']=paris_listings['price'].str.replace(',' , '')

paris_listings.head()
```

```python
paris_listings['price']=paris_listings['price'].astype('float')
mean= paris_listings.iloc[0:5]['price'].mean()
mean
```
